```python
import warnings as wr
wr.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dhaka Rent.csv")
df
```

```python
df.info()
```

```python
plt.scatter(df.area,df.rent,color = "red",marker="*")
plt.title("Dhaka Area-Rent",color = "green")
plt.xlabel("Area",color = "green")
plt.ylabel("Rent",color = "green")
plt.savefig("S1")
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
reg = LinearRegression()
```

```python
reg.fit(df[["area"]],df[["rent"]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(df[["area"]])
```

```python
df["Predicted_Rent"]=reg.predict(df[["area"]])
df
```

```python
plt.plot(df.area,reg.predict(df[["area"]]),color = "blue")
plt.title("Dhaka Area-Rent",color = "green")
plt.xlabel("Area",color = "green")
plt.ylabel("Rent",color = "green")
plt.savefig("S2")
```

```python
plt.scatter(df.area,df.rent,color = "red",marker="*")
plt.plot(df.area,reg.predict(df[["area"]]),color = "blue")
plt.title("Dhaka Area-Rent",color = "green")
plt.xlabel("Area",color = "Green")
plt.ylabel("Rent",color = "green")
plt.savefig("S3")
```

```python
reg.score(df[["area"]],df[["rent"]])
```
