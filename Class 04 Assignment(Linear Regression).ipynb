```python
import warnings as wr
wr.filterwarnings("ignore") # unnecessary item not show
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Height-Weight Data.csv") # csv file import 
```

```python
df.head() # csv file first 5 item show
```

```python
df.info() #csv file informetion
```

```python
plt.scatter(df.Height,df.Weight,color = "red",marker = ".")# Height & Weight graph show
# x axis to height
# y axis to weight
plt.title("Height-Weight Data",color = "blue")#graph title
plt.xlabel("Height",color = "blue")#graph  x axis to title
plt.ylabel("Weight",color = "blue")#graph y axis to title
plt.savefig("img01.png")# graph png file download
```

```python
df.Height.mean() # Height average
```

```python
df.Weight.mean() # Weight average
```

```python
x = df[["Height"]] # Height 2 dimensional array
```

```python
x.head() # csv file Height column 5 item show
```

```python
y = df[["Weight"]] #Weight 2 dimensional array
```

```python
y.head() # csv file Weight column 5 item show
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() #Creating object for LinearRegression 
```

```python
reg.fit(x,y) #teaining the model
```

```python
reg.coef_ #m
```

```python
reg.intercept_ #c
```

```python
reg.predict([[70]]) # any number prediction
```

```python
reg.predict(x) #csv file all number prediction
```

```python
df["Predict_Weight"] = reg.predict(x) # add to column all number prediction
```

```python
df.head()
```

```python
plt.plot(df.Height,reg.predict(df[['Height']]),color = "black") #bast feet line show to graph
# x = Height
# y = Predict Weight
plt.title("Height - Weight Dta",color = "blue")
plt.xlabel("Height",color = "blue")
plt.ylabel("Weight",color = "blue")
plt.savefig("img02.png")
```

```python
plt.plot(df.Height,reg.predict(x),color = "black") 
plt.title("Height - Weight Dta",color = "blue")
plt.xlabel("Height",color = "blue")
plt.ylabel("Weight",color = "blue")
```

```python
plt.plot(df.Height,reg.predict(df[['Height']]),color = "black") #bast feet line show to graph
plt.scatter(df.Height,df.Weight,color = "red",marker="*")## Height & Weight graph show
plt.title("Height - Weight Dta",color = "blue")
plt.xlabel("Height",color = "blue")
plt.ylabel("Weight",color = "blue")
plt.savefig("img03.png")
```

```python
reg.score(x,y) # training score
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df[["Height"]],reg.predict(df[['Height']])) #count of error
```

```python
mean_squared_error(x,reg.predict(x))
```
