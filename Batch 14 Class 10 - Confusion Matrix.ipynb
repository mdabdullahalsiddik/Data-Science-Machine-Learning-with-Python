# Encoder

```python
from sklearn.preprocessing import LabelEncoder # input LabelEncoder
```

```python
le = LabelEncoder()# createing LabelEncoder object
```

```python
from pandas.core.dtypes.common import is_numeric_dtype # input numeric data type
```

```python
# useing  for loop to Encoder
for col in df.columns:
    if is_numeric_dtype(df[col]):
        continue
    else:
        df[col] = le.fit_transform(df[col])
```

# training & testing data

```python
from sklearn.model_selection import train_test_split # inport train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size= 0.70 ,random_state=1) # training data 70 % & testing data 30%
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier #input DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()  # createing LabelEncoder object
```

```python
dtc.fit(xtrain,ytrain)#training and testing data the model
```

```python
dtc.predict(xtest)# all number prediction
```

# Confusion Matrix

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,plot_roc_curve # input classification_report,accuracy_score,confusion_matrix
```

```python
classification_report(ytest,dtc.predict(xtest))
```

```python
print(classification_report(ytest,dtc.predict(xtest))) # accuracy 
```

```python
confusion_matrix(ytest,dtc.predict(xtest)) 
```

```python
 true_positive,false_negative,false_positive,true_negative = confusion_matrix(ytest,dtc.predict(xtest)).reshape(-1) 
```

```python
true_positive
```

```python
false_negative
```

# Accuracy

```python
accuracy = (true_positive + true_negative)/(true_positive + true_negative + false_positive + false_negative)
```

```python
accuracy
```

# Positive Predictive Value

```python
ppv = true_positive / (true_positive + false_positive)
```

```python
ppv
```

# Sensitivity or Recall or Hit Rate or True Positive Rate

```python
tpr = true_positive / (true_positive + false_negative)
```

```python
tpr
```

# False Positive Rate  or 1-Specificity

```python
fpr = false_positive / (false_positive + true_negative)
```

```python
fpr
```

# Specificity or Selectivity or True Negative Rate 

```python
tnr = true_negative / (true_negative + false_positive)
```

```python
tnr
```

# Threat Score  or Critical Success Index

```python
csi = true_positive / (true_positive + false_negative + false_positive)
```

```python
csi
```

# False Discovery Rate

```python
fdr = false_positive / (true_positive + false_positive)
```

```python
fdr
```
